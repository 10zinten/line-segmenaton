```python
import glob

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def plot(img, cmap=None):
    plt.figure(figsize=(20,10))
    plt.axis('off')
    plt.imshow(img, cmap=cmap)
    plt.show()
```

```python
def plot_seg(fns):
    for line, fn in enumerate(fns):
        img = Image.open(fn)
        print("line:", line+1)
        plot(img, cmap='gray')
```
