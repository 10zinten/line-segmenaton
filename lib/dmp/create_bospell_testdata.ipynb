```python
from IPython.display import Markdown, display

def highlight_char(text, idx, color='red'):
    char_list = list(text)
    color_char = "<span style='color:{}'>{}</span>".format(color, text[idx])
    char_list[idx] = color_char
    display(Markdown(''.join(char_list)))
```

```python
def display_mismatch(text_bf, text_af):
    mismatch_indices = [i for i in range(len(text_af)) if text_bf[i] != text_af[i]]
    for idx in mismatch_indices:
        print(idx, text_bf[idx], '-->', text_af[idx])
        
        print('Before:')
        highlight_char(text_bf, idx)
        print('After:')
        highlight_char(text_af, idx)
```

```python
display_mismatch(before_raw_text[:1000], after_raw_text[:1000])
```
