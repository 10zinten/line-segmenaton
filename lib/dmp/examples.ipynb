### Apply the patches on derge-kangyur

```python
# preprocess derge texts
def get_derge_text(lines):
    text_lines = []
    for line in lines:
        if ']' in line and '{' in line:
            btw_text = line[line.find(']')+1: line.find('{')]
            line = line[line.find('}')+1:] if '}' in line else line
            line = btw_text + line if btw_text else line
        else:
            line = line[line.find(']')+1:]
        text_lines.append(line)
    return text_lines
```

```python
def get_ocr_pages(fns):
    avoided = 'འཕགས་པ་དཀོན་མཆོག་ཏ་ལ་ལའི་གཟུངས་ཞེས་བྱ་བ་ཐེག་པ་ཆེན་པོའི་མདོ།'
    pages = []
    for fn in fns:
        lines = []
        for line in get_texts(fn):
            if len(line) > 50 and avoided not in line:
                line = line.replace('“', '')
                line = line.replace('》', '')
                line = line.replace('：', '')
                line = line.replace(' ་ ་ ་ ་', '་་་་')
                lines.append(line)
        pages.append(''.join(lines))
    return pages
```
