```python
import os
import glob

import diff_match_patch as dmp_module
```

```python
def get_texts(fn):
    with open(fn) as f:
        return f.readlines()
```

```python
ocr_texts_path = '../../data/book-97/texts/'
derge_kangyur_path = '../../../derge-kangyur/derge-kangyur-tags/100-tagged.txt' # book-97 == derge-kangyur-tag-100
```

```python
ocr_texts_fns = sorted(glob.glob(os.path.join(ocr_texts_path, "*")))
ocr_texts_fns
```

```python
ocr_text_39 = get_texts(ocr_texts_fns[0]); ocr_text_39
```

```python
derge_text = get_texts(derge_kangyur_path)[:8]
derge_text
```

### DMP diff the ocr text with new-lines and without new-lines
- Preprocessing: remove all the uncessary line - with line lenght

```python
ocr_texts_w_nl = ''.join([line for line in ocr_text_39 if len(line) > 50])
ocr_texts_w_nl
```

```python
ocr_texts_wo_nl = ''.join([line.rstrip('\n') for line in ocr_texts_w_nl])
ocr_texts_wo_nl
```

```python
dmp = dmp_module.diff_match_patch()
diff = dmp.diff_main(ocr_texts_wo_nl, ocr_texts_w_nl)
diff
```

### Create patches

```python
patches = dmp.patch_make(diff)
```

### Apply the patches on derge-kangyur

```python
# preprocess derge texts
def get_derge_text(lines):
    text_lines = []
    for line in lines[3:]:
        line = line[line.find(']')+1:]
        line = line[line.find('}')+1:] if '}' in line else line
        text_lines.append(line)
    return text_lines
```

```python
derge_text_prep = get_derge_text(derge_text)
derge_text_prep
```

```python
derge_text_prep = ''.join([line.rstrip('\n') for line in derge_text_prep])
derge_text_prep
```

```python
output = dmp.patch_apply(patches, derge_text_prep)
output
```
